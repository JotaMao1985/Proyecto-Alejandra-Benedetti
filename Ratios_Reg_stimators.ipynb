```python
import pandas as pd
import numpy as np

# Cargar datos
# Nota: El archivo agsrs.csv está en la carpeta "CSV data sets for SDA 3e"
df = pd.read_csv('CSV data sets for SDA 3e/agsrs.csv')

# Variables
y = df['acres92']  # Variable de interés
x = df['acres87']  # Variable auxiliar

# Estadísticos descriptivos
print("=== Estadísticos Descriptivos ===")
print(f"n = {len(y)} condados en la muestra")
print(f"\nacres92 (y):")
print(f"  Suma: {y.sum():,.0f}")
print(f"  Media: {y.mean():,.2f}")
print(f"  Desviación estándar: {y.std():,.2f}")
print(f"\nacres87 (x):")
print(f"  Suma: {x.sum():,.0f}")
print(f"  Media: {x.mean():,.2f}")
print(f"  Desviación estándar: {x.std():,.2f}")

# Ratio estimado
B_hat = y.sum() / x.sum()
print(f"\n=== Estimador de Ratio ===")
print(f"Ratio estimado B_hat = {B_hat:.6f}")

# Total poblacional conocido de x
t_x = 964470622  # Total de acres en 1987 para N=3041 condados

# Estimador de ratio para el total
t_y_ratio = B_hat * t_x
print(f"Total estimado (ratio): {t_y_ratio:,.0f} acres")

# Correlación
r = np.corrcoef(y, x)[0, 1]
print(f"Correlación entre acres92 y acres87: {r:.4f}")

# Varianza del estimador de ratio
n = len(y)
N = 3041  # Total de condados en la población
f = n / N  # Fracción de muestreo

# Sumas de cuadrados
s_x2 = x.var(ddof=1)
s_y2 = y.var(ddof=1)
s_xy = x.cov(y)

# Varianza aproximada del ratio
var_B = (1 - f) * (1 / (n * x.mean()**2)) * (s_y2 - 2 * B_hat * s_xy + B_hat**2 * s_x2)
EE_B = np.sqrt(var_B)
print(f"\nError estándar del ratio: {EE_B:.6f}")

# Error estándar del total estimado
EE_total = t_x * EE_B
print(f"Error estándar del total estimado: {EE_total:,.0f} acres")

# Estimador directo (expansión simple)
t_y_directo = N * y.mean()
var_directo = N**2 * (1 - f) * s_y2 / n
EE_directo = np.sqrt(var_directo)

print(f"\n=== Comparación con Estimador Directo ===")
print(f"Total estimado (directo): {t_y_directo:,.0f} acres")
print(f"Error estándar (directo): {EE_directo:,.0f} acres")
print(f"Error estándar (ratio): {EE_total:,.0f} acres")
print(f"Reducción en varianza: {(1 - var_B/var_directo)*100:.1f}%")

# Análisis de regresión
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print(f"\n=== Análisis de Regresión ===")
print(f"Pendiente (b): {slope:.6f}")
print(f"Intercepto (a): {intercept:.2f}")
print(f"R²: {r_value**2:.4f}")

# Estimador de regresión para el total
mu_x = t_x / N  # Media poblacional de x
y_bar = y.mean()
x_bar = x.mean()

# Estimador de regresión de la media
y_bar_reg = y_bar + slope * (mu_x - x_bar)
t_y_reg = N * y_bar_reg

print(f"\n=== Estimador de Regresión ===")
print(f"Media poblacional de x (mu_x): {mu_x:,.2f}")
print(f"Media muestral de x (x_bar): {x_bar:,.2f}")
print(f"Media estimada por regresión: {y_bar_reg:,.2f}")
print(f"Total estimado (regresión): {t_y_reg:,.0f} acres")
```
